```python
merged_data1 = pd.merge(e_items, e_orders, on='item_id')
merged_data2 = pd.merge(merged_data1, e_sales, on='sales_id')
merged_data3 = pd.merge(merged_data2, e_customers, on='customer_id')
```

```python
filtered_data = merged_data3[(merged_data3['age'] >= 18) & (merged_data3['age'] <= 35)]
```

```python
grouped_data = filtered_data.groupby(['customer_id','item_name','item_id']).agg(total_quantity=('quantity', 'sum')).reset_index()
```
